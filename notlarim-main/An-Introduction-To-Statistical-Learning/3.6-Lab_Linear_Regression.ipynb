`lm(y~x, data)` Fitting a simple linear regression model.

`y` response, `x` predictor, `data` data set which these two variables are kept.

```r
attach(Boston)
lm.fit = lm(medv~lstat)
```

```r
lm.fit
```

```r
summary(lm.fit)
```

```r
names(lm.fit)
```

`coef()` function to access coefficient values.

```r
coef(lm.fit)
```

`confint()` function to access confidence interval for the coefficient estimates.

```r
confint(lm.fit)
```

`predict()` function to produce **confidence intervals** and **prediction intervals** for the prediction of _medv_ for a given value of _lstat_.

```r
predict(lm.fit, data.frame(lstat=c(5,10,15)),
        interval="confidence")
```
